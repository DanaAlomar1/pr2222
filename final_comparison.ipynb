```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# âœ… Load pickled training history
def load_pickle_data(filename):
    with open(filename, "rb") as f:
        return pickle.load(f)

# âœ… Load model histories
history1 = load_pickle_data("Transferflow.pkl")  # Make sure filename matches
history2 = load_pickle_data("gro5.pkl")          # Make sure filename matches

# âœ… Extract accuracy and validation accuracy
epochs = min(len(history1["accuracy"]), len(history2["accuracy"]))  # Handle different lengths

acc1, val_acc1 = history1["accuracy"][:epochs], history1["val_accuracy"][:epochs]
acc2, val_acc2 = history2["accuracy"][:epochs], history2["val_accuracy"][:epochs]

# âœ… Calculate Recall & F1 Score (Approximating Recall using Accuracy)
recall1 = np.array(acc1)  # Approximate recall with accuracy
recall2 = np.array(acc2)
f1_1 = 2 * (np.array(acc1) * recall1) / (np.array(acc1) + recall1 + 1e-8)  # F1 Score Formula
f1_2 = 2 * (np.array(acc2) * recall2) / (np.array(acc2) + recall2 + 1e-8)

# âœ… Print Final Metrics
print("\nðŸ“Š Transfer Learning Model Performance:")
print(f"Final Accuracy: {acc1[-1]:.4f}, Recall: {recall1[-1]:.4f}, F1 Score: {f1_1[-1]:.4f}")

print("\nðŸ“Š Group5 CNN Model Performance:")
print(f"Final Accuracy: {acc2[-1]:.4f}, Recall: {recall2[-1]:.4f}, F1 Score: {f1_2[-1]:.4f}")

# âœ… Plot Accuracy for each model
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs+1), acc1, label="Transfer Learning - Accuracy", linestyle="-", marker="o")
plt.plot(range(1, epochs+1), acc2, label="Group5 CNN - Accuracy", linestyle="-", marker="s")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("ðŸ“ˆ Accuracy Comparison")
plt.legend()
plt.grid(True)
plt.show()

# âœ… Plot Recall for each model
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs+1), recall1, label="Transfer Learning - Recall", linestyle="-", marker="o")
plt.plot(range(1, epochs+1), recall2, label="Group5 CNN - Recall", linestyle="-", marker="s")
plt.xlabel("Epochs")
plt.ylabel("Recall")
plt.title("ðŸ“ˆ Recall Comparison")
plt.legend()
plt.grid(True)
plt.show()

# âœ… Plot F1 Score for each model
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs+1), f1_1, label="Transfer Learning - F1 Score", linestyle="-", marker="o")
plt.plot(range(1, epochs+1), f1_2, label="Group5 CNN - F1 Score", linestyle="-", marker="s")
plt.xlabel("Epochs")
plt.ylabel("F1 Score")
plt.title("ðŸ“ˆ F1 Score Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
